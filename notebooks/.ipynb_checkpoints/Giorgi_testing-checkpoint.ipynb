```python
client = bigquery.Client()
query = """
# Retrieves all events with click actions (event_action = select_content),their queries, resulting products, and number of queries resulting in specific product
# Retrieves all events with click actions (event_action = select_content),their queries, and resulting products for the first PLP

WITH session_target AS (
  SELECT
    hit_sequence_number,
    event_category,
    event_action,
    session_id,
    page_url_query_parameter,
    next_page_url,
    SPLIT(REGEXP_EXTRACT(page_url_query_parameter, r'\?q=(\w.+)+'), '&')[SAFE_OFFSET(0)] AS clean_query,
    ARRAY_REVERSE(SPLIT(next_page_url,'/'))[SAFE_OFFSET(0)] AS target_product,
    ARRAY_REVERSE(SPLIT(ARRAY_REVERSE(SPLIT(next_page_url,'/'))[SAFE_OFFSET(0)], '-'))[SAFE_OFFSET(0)] AS target_product_id,
  FROM
    ingka-web-analytics-prod.web_data_v2.hits_events_and_pages
  WHERE
    ( date_hit BETWEEN DATE_SUB(CURRENT_DATE(), INTERVAL 365 DAY) AND CURRENT_DATE()
      AND website_market_short = 'us'
      AND event_category = 'engagement'
      AND event_action = 'select_content'
      AND page_url_query_parameter IS NOT NULL) 

),
clean_query_product AS (
  SELECT
    session_id,
    clean_query,
    REGEXP_REPLACE(target_product, r'-', ' ') AS target_product,
    target_product_id,
    COUNT(*) AS query_n_resulting_product
  FROM
    session_target
  WHERE
    clean_query IS NOT NULL
    AND REGEXP_CONTAINS(target_product, r'[0-9A-Za-z]*[\d]')--target_product_id NOT IN ('products', 'en', 'Site Exit', 'gallery', 'shoppingcart', 'latest', 'search', 'login') -- prevent the target_product to become one of these 'next_pages'
  GROUP BY
    clean_query,
    target_product_id,
    target_product,
    session_id
  ORDER BY
    clean_query
)

SELECT 
  clean_query, 
  ARRAY_AGG(target_product_id IGNORE NULLS order by query_n_resulting_product desc) as term_frequency_ranking
FROM clean_query_product 
GROUP BY clean_query

"""
r = client.query(query).to_dataframe()
```

```python
client = bigquery.Client()
query = """
  # Retrieves all events with click actions (event_action = select_content),their queries, and resulting products for the first PLP
  
WITH session_counts AS(
  SELECT
  SPLIT(REGEXP_EXTRACT(page_url_query_parameter, r'\?q=(\w.+)+'), '&')[SAFE_OFFSET(0)] AS clean_query,
  ARRAY_REVERSE(SPLIT(next_page_url,'/'))[SAFE_OFFSET(0)] AS target_product,
  ARRAY_REVERSE(SPLIT(ARRAY_REVERSE(SPLIT(next_page_url,'/'))[SAFE_OFFSET(0)], '-'))[SAFE_OFFSET(0)] AS target_product_id,

FROM
  ingka-web-analytics-prod.web_data_v2.hits_events_and_pages 

WHERE
  ( 
    date_hit between DATE_SUB(current_date(), INTERVAL 365 DAY) and current_date()
    AND website_market_short = 'us'
    AND event_category = 'engagement'
    AND event_action = 'select_content'
    AND page_url_query_parameter IS NOT NULL
  )
),

clean_query_product AS (
SELECT 
  clean_query, 
  REGEXP_REPLACE(target_product, r'-', ' ') as target_product, 
  target_product_id, COUNT(*) as n_sessions_product_resulting_query
FROM session_counts 
WHERE 
    clean_query IS NOT NULL

GROUP BY
  clean_query, target_product_id, target_product

ORDER BY
  clean_query
)

SELECT 
  target_product_id, 
  ARRAY_AGG((clean_query, n_sessions_product_resulting_query) IGNORE NULLS order by n_sessions_product_resulting_query desc) as term_frequency_ranking
FROM clean_query_product 
GROUP BY target_product_id


"""
df = client.query(query).to_dataframe()
```

```python
def is_item_no(s):
    return np.array([i.isdigit() for i in s]).all()

def add_or(li, sep=','):
    return f'{sep} '.join(li)
    
    

def create_option_string(df, column, inds):
    options = ""
    for i in inds:
        options +=  f"'{df[column].iloc[i]}',"
        #if i == inds[-1]:
        #    options = options[:-1]
            
    
    option_list = options.replace("'", "").split(',')[:-1]
    option_string = add_or(option_list)
    
    return option_string, option_list

def get_other_indices(index, size, num):
    """
    given a number {index}, returns {num} integers in range({size})
    """
    
    proceed = 0
    while proceed == 0:
        inds = np.random.choice(range(size), num)
        # check if any of the new indices is same as initial
        if (inds == index).any()==False:
            proceed = 1
        else:
            proceed = 0
    
    return inds

def hierarchy_prompt(hierarchy_df, item_no=None):
    w = 0    
    if item_no is not None:
        row = hierarchy_df[hierarchy_df['item_no']==item_no]
        while len(row['triplet'].iloc[0]) < 1:
            row = hierarchy_df.sample()
            w+=1
            if w > 100:
                print("help")
        item_no = row['item_no'].iloc[0]
    else:
        row = hierarchy_df.sample()
        while len(row['triplet'].iloc[0]) < 1:
            row = hierarchy_df.sample()
            w+=1
            if w > 100:
                print("help")
        item_no = row['item_no'].iloc[0]
    
    triplets = row['triplet'].iloc[0]
    triplet = triplets[np.random.randint(len(triplets))]
    b=0
    while 'PLACEHOLDER' in triplet:
        b+=1
        triplet = triplets[np.random.randint(len(triplets))]
        if b>100:
            print("hierarchy stuck in loop!")
            return None, None
    
    target = triplet
    triplist = triplet.split(',')
    np.random.shuffle(triplist)
    inptrip = ', '.join(triplist)

    np.random.shuffle(triplist)
    permutations = list(itertools.permutations(triplist))
    option_list = [', '.join(t) for t in permutations]
    options = add_or(option_list, ';')

    inp = f"out of the following options: {inptrip} what  are properties of item_no: {item_no} from most specific to least specific? " + options
    
    return inp.lower(), target#, option_list

def true_query(r, item_no):
    
    is_true = np.random.randint(2)
    
    if is_true:
        query = r[r['item_no']==item_no].iloc[0,np.random.randint(1,21)]
        target = "Yes"
    else:
        new_r = r[r!=item_no].dropna()
        query = new_r.iloc[np.random.randint(len(new_r)),np.random.randint(1, 10)]
        target = "No"

    
    # create input
    inp = query.lower() + f".\ndoes the sentence above describe item_no {item_no}? Yes or No?"
    
    return inp, target#, ['Yes', 'No']

def why_similar(queries, query_non_singular, n_opts):
        
    row = query_non_singular.sample()
    query = row['clean_query'].iloc[0]
    b = 0
    while 'PLACEHOLDER'==query:
        row = query_non_singular.sample()
        query = row['clean_query'].iloc[0]
        if b>100:
            print("similar stuck in placeholder loop!")
    clean_row = np.unique(row[l].values[row[l].values != None])
    ids = clean_row[sample(range(0, len(clean_row)), 2)]
    b = 0
    while ids[0]==ids[1]:
        ids = clean_row[sample(range(0, len(clean_row)), 2)]
        b+=1
        if b>100:
            print(clean_row)
            print("why sim stuck in loop!")
    target = f"{query} describes {ids[0]} and {ids[1]}"
    option_list = queries[(queries!=ids[0])&(queries!=ids[1])].dropna()['clean_query'].sample(n_opts).tolist() + [query]
    np.random.shuffle(option_list)
    
    # get options as a string
    options = ', '.join([f"{q} describes {ids[0]} and {ids[1]}" for q in option_list])
    
    # create input
    inp = f"item_no {ids[0]} and {ids[1]} are similar, why? Which of the following is the correct answer? " + options
    
    return inp.lower(), target#, option_list

def true_similar(queries, query_non_singular, item_no):
    
    if np.random.random()>0.5:
        row = query_non_singular.sample()
        query = row['clean_query'].iloc[0]
        clean_row = np.unique(row[l].values[row[l].values != None])
        ids = clean_row[sample(range(0, len(clean_row)), 2)]
        b = 0
        while ids[0]==ids[1]:
            ids = clean_row[sample(range(0, len(clean_row)), 2)]
            b+=1
            if b>100:
                print(clean_row)
                print("true sim stuck in loop!")
                break
        target = "True"
    else:
        filtered_queries = queries[queries!=item_no][l].values
        filtered_queries = pd.Series(filtered_queries[filtered_queries!=None]).sample()
        ids = np.array([filtered_queries.iloc[0], item_no])
        np.random.shuffle(ids)
        target = "False"
    

    # create input
    inp = f"{ids[0]} and {ids[1]} are similar. True or False?"
    
    return inp, target#, []

def query_top_3_result(query_3_result):
        
    # get random query
    row = query_3_result.sample()
    query = row['clean_query'].iloc[0]
    top3 = list(row[['rank_1', 'rank_2', 'rank_3']].values[0])
    top3_permutations = list(itertools.permutations(top3))
    target = ', '.join(top3)
    option_list = [', '.join(t) for t in top3_permutations]
    np.random.shuffle(option_list)
    
    # get options as a string
    options = add_or(option_list, ';')
    
    # create input
    inp = "query: '" + query + f"'\nfor the query above, what are the top 3 item_no that best match in the correct order? " + options + '?' 
    
    return inp.lower(), target#, option_list


def query_top_result(query_df, r, item_no, n_opts):
        
    # get random query
    query = query_df['clean_query'].sample().iloc[0]
    while 'PLACEHOLDER'==query:
        query = query_df['clean_query'].sample().iloc[0]
        if b>100:
            print("similar stuck in placeholder loop!")
    
    # get random item from items who do not share any of the top queries
    option_list = r[r!=query].dropna()['item_no'].sample(n_opts).tolist()+[item_no]
    np.random.shuffle(option_list)
    
    # get answer options
    options = add_or(option_list) if np.random.random()>0.02 else ""

    target = item_no
    
    # create input
    inp = "query: '" + query + f"'\nfor the query above, what item_no is the correct match? " + options + '?' 
    
    return inp.lower(), target#, option_list

                           
def item_top_query(r, item_no, n_opts):
    
    tar = r[r['item_no']==item_no].iloc[0,np.random.randint(1,21)]
    while 'PLACEHOLDER'==tar:
        tar = r[r['item_no']==item_no].iloc[0,np.random.randint(1,21)]
        if b>100:
            print("item_top_query stuck in placeholder loop!")
                           
    # get other items
    new_r = r[r!=tar].dropna()
    option_list = [new_r.iloc[np.random.randint(len(new_r)),np.random.randint(1, 21)] for i in range(n_opts)] + [tar]
    np.random.shuffle(option_list)

    # get answer options
    options = ', '.join([f"{q} describes {item_no}" for q in option_list])


    inp = "for item_no: '" + item_no + "'. Which of the following is the best match? " + options + '?'
    target = f"{tar} describes {item_no}"

    return inp.lower(), target#, option_list
```
